1. Import libraries

```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt 
import seaborn as sns
```

2. Load datasets

```python
sessions = pd.read_csv('../../data/website_sessions.csv')
pageviews = pd.read_csv('../../data/website_pageviews.csv')
orders = pd.read_csv('../../data/orders.csv')
order_items = pd.read_csv('../../data/order_items.csv')
refunds = pd.read_csv('../../data/order_item_refunds.csv')
products = pd.read_csv('../../data/products.csv')
```

3. Display shape

```python
pd.DataFrame({
    'sessions': sessions.shape,
    'pageviews': pageviews.shape,
    'orders': orders.shape,
    'order_items': order_items.shape,
    'refunds': refunds.shape,
    'products': products.shape
}).T.rename(columns={0: 'rows', 1: 'columns'})
```

```python
pd.DataFrame({
    "table": [
        "sessions",
        "pageviews",
        "orders",
        "order_items",
        "refunds",
        "products"
    ],
    "rows": [
        sessions.shape[0],
        pageviews.shape[0],
        orders.shape[0],
        order_items.shape[0],
        refunds.shape[0],
        products.shape[0]
    ],
    "columns": [
        sessions.shape[1],
        pageviews.shape[1],
        orders.shape[1],
        order_items.shape[1],
        refunds.shape[1],
        products.shape[1]
    ]
})
```

7. Convert Datetime

```python
sessions['created_at'] = pd.to_datetime(sessions['created_at'])
pageviews['created_at'] = pd.to_datetime(pageviews['created_at'])
orders['created_at'] = pd.to_datetime(orders['created_at'])
order_items['created_at'] = pd.to_datetime(order_items['created_at'])
refunds['created_at'] = pd.to_datetime(refunds['created_at'])
products['created_at'] = pd.to_datetime(products['created_at'])
```

8. Create Analytical Datasets

8.1. Session -> Order

```python
session_order = sessions.merge(orders, on='website_session_id', how='left')
```

```python
session_order.shape
```

```python
session_order.head()
```

8.2. Order -> Product

```python
order_product = orders.merge(order_items, on='order_id', how='left').merge(products, on='product_id', how='left')
```

```python
order_product.shape
```

```python
order_product.head()
```

8.3. Order -> Refund

```python
order_refund = order_items.merge(refunds, on='order_item_id', how='left')
```
